## Тема 3: Построение модели классификации.

__1.__ Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted? 

Разные варианты усреднения нужны для оценки работы модели на выборках с учетом баланса классов. **Macro-average** - подсчет precision/recall по всем классам отдельно и затем усреднение полученных результатов без учета дисбаланса классов.

При подсчете **weighted-average** берется взвешенное среднее метрик по всем классам, т.е. учитывается вес каждого класса в общей выборке в зависимости от его представленности.

**Micro-average** оптимален в случае, если классов больше 2, и делает оценку в зависимости от совокупного по всем классам количества TP, FP и FN. Micro-average precision рассчитывается как сумма TP по обоим классам, деленная на сумму TP и FP по обоим классам. Micro-average recall рассчитывается как сумма TP по обоим классам, деленная на сумму TP и FN по обоим классам.

__2.__ В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?


Суть алгоритма:
- catboost: градиентный бустинг деревьев решений, преимущество - возможность лучше обрабатывать кат. фичи.
- lightgbm: для разделения классов используется метод односторонней выборки на основе градиента (goss), преимущество - скорость работы алгоритма на числовых признаках;
- xgboost: для разделения классов используется алгоритм на основе гистограммы, самый старый из этих алгоритмов.

Работа с категоривальными признаками:
- catboost может обрабатывать категориальные фичи при помощи кодирования;
- lightgbm может обрабатывать, но только в формате int;
- xgboost не обрабатывает кат. признаки.


Гиперпараметры:
Т.к. все алгоритмы представляют собой ансамбли деревьев, у всех присутствует параметры глубины max-depth/depth, которые контролируют переобучение. В catboost есть параметр L2-регуляризации, в отличие от других моделей.


## Тема 2: Анализ данных и проверка статистических гипотез.

__1.__ В чём различие между зависимыми и независимыми выборками?


Зависимые выборки - это выборки, связанные друг с другом. Например, сравнение одной и той же выборки людей до и после теста лекарства, или выборки сотрудников одной компании.
Независимые выборки - выборки, элементы которых не связаны друг с другом и выбраны из разных генеральных совокупностей (до определенного уровня). Например, мужчины и женщины, жители разных стран, сотрудники разных компаний.

Важно использование выборок правильного типа для каждого исследования. Если мы хотим изучить динамику, то нужно брать зависимые выборки. Если необходимо провести независимое обобщающее ислледование, нужно использовать независимые выборки. Например, если мы исследуем влияние лекарства на выборках, в которых есть близнецы, т.е. у участников схожи анатомические признаки и показатели здоровья, то мы не получим достоверных результатов. Также примером зависимых выборок являются выборки сотрудников одной компании в случае исследования удовлетворонности рабочими условиями в стране.

__2.__ Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

При проверке гипотезы, если мы знаем распределение выборки/выборок, то мы используем параметрические критерии, если распредление неизвестно - то непараметрические. Например, мы можем провести проверку двух выборок на нормальность (Критерий Шапиро-Уолка) и если p-value будет больше $alpha$ (т.е. распределение признано нормальным), то мы сможем применять параметрические критерии (например, t-тест), если гипотеза о нормальности выборок отвергнута, можем применить непараметрический U-критерий Манна-Уитни.

## Тема 1: Метрики для задачи классификации: precision, recall, F-score

__1.__ Привести по 2 примера, когда лучше максимизировать Precision, а когда Recall?

precision лучше оптимизировать, когда нам нужно оценить насколько мы правильно предсказываем целевой класс. Это значит, что повышение этой метрики актуально для задач, в которых ошибочное отнесение к целевому классу (ошибка первого рода/ ложная тревога/ false positive) нежелательно. Например:

- задача классификации письма (определение спама). важное письмо попадает в спам и не доходит до получателя;

- задача классификации пациента (наличие заболевания). неправильная постановка серьезного диагноза может привести к тому, что пациент подвергнется опасному хирургическому вмешательству, которое может стоить ему жизни.


recall нужно повышать, когда нам нужно максимизировать долю истинных объектов, которых мы относим к целевому классу. То есть нам важно покрыть все истинные объекты нашей моделью, и ошибка второго рода недопустима (false negative), т.к. может стоить очень дорого для компании или повлечь за собой опасность для человека. Например:

- задача классификации пилотов (допуск к полету);
- задача классификации воздушного судна (допуск к рейсу);
- задача классификации клиента банка (выплатит кредит). 

__2.__ Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?


F1-метрика оптимальна, т.к. позволяет принять решение на основе одного числа, а не двух.

Взять среднее арифметическое по recall и precision не является оптимальным решением, т.к. во многих задачах нужен вес, определяющий важность recall или precision (во избежание ошибок 1 или 2 рода).

В случае, если мы не взвешиваем recall и precision, гармоническое среднее, которым является F1, позволяет избежать завышенности метрики, т.к. оно менее чувствительно к большим выбросам и ориентировано на наименьший элемент в списке.

```python
recall = 0.3
precision = 0.8


avg = (recall+precision)/2
F1 = 2*recall*precision/(recall+precision)

print(f'Recall and precision average:\t{avg:.2f}')
print(f'F1-score:\t\t\t{F1:.2f}')
```

При использовании F1-score мы получаем не завышенную метрику, что позволяет более адекватно оценивать модель.
